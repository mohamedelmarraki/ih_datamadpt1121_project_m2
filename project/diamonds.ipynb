```python
import pandas as pd
import sqlite3 as sql
import sqlalchemy
```

```python
connection = sql.connect("diamonds_m2.db")
```

```python
print(pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", connection))
```

```python
query1= 'SELECT * FROM diamonds_dimensions'
df_dimensions = pd.read_sql_query(query1, connection)
df_dimensions
```

```python
query2= 'SELECT * FROM diamonds_transactional'
df_transactional = pd.read_sql_query(query2, connection)
df_transactional
```

```python
query3= 'SELECT * FROM diamonds_properties'
df_properties = pd.read_sql_query(query3, connection)
df_properties
```

```python
query4= 'SELECT * FROM diamonds_cut'
df_cut = pd.read_sql_query(query4, connection)
df_cut
```

```python
query5= 'SELECT * FROM diamonds_color'
df_color = pd.read_sql_query(query5, connection)
lista = []
for i in df_color["color"]:
    if i == "D":
        lista.append("Blanco excepcional +")
    if i == "E":
        lista.append("Blanco excepcional")
    if i == "F":
        lista.append("Blanco raro +")
    if i == "G":
        lista.append("Blanco raro")
    if i == "H":
        lista.append("Blanco")
    if i == "I":
        lista.append("Blanco ligeramente coloreado")
    if i == "J":
        lista.append("Blanco ligeramente coloreado")
df_color["Color_Explicación"]=lista
df_color
```

```python
query6= 'SELECT * FROM diamonds_clarity'
df_clarity = pd.read_sql_query(query6, connection)
lista1 = []
for i in df_clarity["clarity"]:
    if i == "SI2":
        lista1.append("pequeña inclusión 2")
    if i == "VVS2":
        lista1.append("Muy muy pequeñas inclusiones 2")
    if i == "VVS1":
        lista1.append("Muy muy pequeñas inclusiones 1")
    if i == "SI1":
        lista1.append("pequeña inclusión 1")
    if i == "I1":
        lista1.append("imperfecto")
    if i == "VS1":
        lista1.append("Muy pequeña inclusión 1")
    if i == "VS2":
        lista1.append("Muy pequeña inclusión 1")
    if i == "IF":
        lista1.append("Sin inclusiones internas")
        
df_clarity["clarity_Explicación"]=lista1

df_clarity
```

```python
query7= 'SELECT * FROM diamonds_city'
df_city = pd.read_sql_query(query7, connection)
lista3 = []
for i in df_city["city"]:
    if i == "Amsterdam":
        lista3.append("Netherlands")
    if i == "Zurich":
        lista3.append("Suiza")
    if i == "Las Vegas":
        lista3.append("Estados Unidos")
    if i == "New York City":
        lista3.append("Estados unidos")
    if i == "Paris":
        lista3.append("Francia")
    if i == "Tel Aviv":
        lista3.append("Palestina")
    if i == "Dubai":
        lista3.append("Emiratos Árabes Unidos")
    if i == "Kimberly":
        lista3.append("Sudáfrica")
    if i == "Surat":
        lista3.append("India")
    if i == "Antwerp":
        lista3.append("Belgica")
    if i == "Madrid":
        lista3.append("España")
    if i == "London":
        lista3.append("Reino unido")
    if i == "Luxembourg":
        lista3.append("Luxemburgo")
df_city["País"]=lista3
df_city
```

```python
join1 = df_dimensions.merge(df_transactional , on = "index_id")
join2 = join1.merge(df_properties , on = "index_id")
join3 = join2.merge(df_cut , on = "cut_id")
join4 = join3.merge(df_color, on = "color_id")
join5 = join4.merge(df_clarity, on = "clarity_id")
join6 = join5.merge(df_city, on = "city_id")
```

```python
join6
```

```python
df_final = join6.loc[:, ["index_id", 'depth', 'table', 'x', 'y', 'z', 'price', 'cut', 'color', "Color_Explicación", 'clarity',"clarity_Explicación", 'city',"País","carat"]]
df_final.info()
```
